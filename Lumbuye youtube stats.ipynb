```python
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
driver.get("https://www.youtube.com/c/LumbuyeFred/videos")
```

```python
driver.find_element_by_tag_name("body").send_keys(Keys.END)
```

```python
for _ in range(50):
    driver.find_element_by_tag_name("body").send_keys(Keys.END)
    time.sleep(3)
```

```python
html = driver.page_source
```

```python
soup = BeautifulSoup(html, "html.parser")
```

```python
videos = soup.find_all("div", {"id": "dismissible"})
```

```python
len(videos)
```

```python
data = []
for video in videos:
    data_dic = {}
    data_dic["title"] = video.find("a", {"id": "video-title"}).text
    data_dic["url"] = "https://www.youtube.com/"+video.find("a", {"id": "video-title"})["href"]
    meta = video.find("div", {"id": "metadata-line"}).find_all("span")
    data_dic["views"] = meta[0].text
    data_dic["date"] = meta[1].text
    data.append(data_dic)
```

```python
data
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.to_csv("lumbuye-stats.csv")
```

```python
def clean_views(df):
    if "K" in df["views"]:
        views = float(df["views"].split("K")[0])*1000
        return views
    
```

```python
df["views_expand"] = df.apply(clean_views, axis=1)
```
